```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/blood-transfusion/transfusion.data")
```

```python
sns.lmplot(x = "Recency (months)", y = "Frequency (times)", hue = 'whether he/she donated blood in March 2007', data = Data)
```

```python
sns.lmplot(x = "Recency (months)", y = "Monetary (c.c. blood)", hue = "whether he/she donated blood in March 2007", data = Data)
```

```python
sns.lmplot(x = "Recency (months)", y = "Time (months)", hue = "whether he/she donated blood in March 2007", data = Data)
```

```python
sns.lmplot(x = "Time (months)", y = "Recency (months)", hue = "whether he/she donated blood in March 2007", data = Data)
```

```python
sns.lmplot(x = "Frequency (times)", y = "Monetary (c.c. blood)", hue = "whether he/she donated blood in March 2007", data = Data)
```

```python
sns.lmplot(x = "Frequency (times)", y = "Time (months)", hue = "whether he/she donated blood in March 2007", data = Data)
```

```python
sns.lmplot(x = "Monetary (c.c. blood)", y = "Time (months)", hue = "whether he/she donated blood in March 2007", data = Data)
```

```python
allPlots = sns.PairGrid(Data, hue = "whether he/she donated blood in March 2007", height = 7)
allPlots.map(sns.regplot)
allPlots.add_legend()
```
